```python
import pandas as pd
```

```python
df_toxic_normal = pd.read_csv("./analysis/data/russian_comments_from_2ch_pikabu.csv")
df_spam = pd.read_csv("./analysis/data/Advert_Spam.csv")
```

```python
df_toxic_normal = df_toxic_normal[['comment', 'toxic']]
df_toxic_normal = df_toxic_normal.rename(columns={"comment":"text"})
```

```python
df_spam = pd.concat([df_spam, pd.Series([1.0 for _ in range(df_spam.shape[0])], name="spam")], axis=1, names=["text", "spam"])
df_spam = df_spam.rename(columns={"message":"text"})
df_spam
```

```python
dataset = pd.concat([df_spam, df_toxic_normal])
dataset.text.isna().value_counts()
```

```python
dataset = dataset.fillna(0.0)
dataset
```

```python
dataset['ok'] = ((dataset['spam'] == 0.0) & (dataset['toxic'] == 0.0)).astype(float)
dataset
```
